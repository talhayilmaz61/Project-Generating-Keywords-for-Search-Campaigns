```python
import pandas as pd

words = ["buy","prices","affordable", "cheap", "good","comfortable"]
products = ["sofas","convertible sofas","love seats","recliners","sofa beds"]

def keyword_df_creator(words, products):
    keyword_list = []
    
    for i in range(len(words)):
        for j in range(len(products)):
            keyword = words[i] + " " + products[j]
            keyword_list.append(keyword)
    for k in range(len(products)):
        for m in range(len(words)):
            keyword_inverse = products[k] + " " + words[m]
            keyword_list.append(keyword_inverse)
    keywords_df = pd.DataFrame(keyword_list, columns=['Keyword'])
    return keywords_df

def keyword_csv_creator(words, products):
    keyword_list = []
    
    for i in range(len(words)):
        for j in range(len(products)):
            keyword = words[i] + " " + products[j]
            keyword_list.append(keyword)
    for k in range(len(products)):
        for m in range(len(words)):
            keyword_inverse = products[k] + " " + words[m]
            keyword_list.append(keyword_inverse)
    keywords_df = pd.DataFrame(keyword_list, columns=['Keyword'])
    keywords_df.to_csv('keywords.csv', index=False)
    return 'keywords.csv'

keywords_df = keyword_df_creator(words, products)
keywords = keyword_csv_creator(words, products)

keywords_df["Ad Group"]=None
keywords_df.loc[:29, "Ad Group"] = products * 6
keywords_df.loc[30:35, "Ad Group"] = products [0]
keywords_df.loc[36:41, "Ad Group"] = products [1]
keywords_df.loc[42:47, "Ad Group"] = products [2]
keywords_df.loc[48:53, "Ad Group"] = products [3]
keywords_df.loc[54:59, "Ad Group"] = products [4]
keywords_df["Campaign"] = "SEM_Sofas"
keywords_df["Criterion Type"] = "Exact"

keywords_df = keywords_df[["Campaign", "Ad Group", 'Keyword',"Criterion Type"]]

print(keywords_df)
```
